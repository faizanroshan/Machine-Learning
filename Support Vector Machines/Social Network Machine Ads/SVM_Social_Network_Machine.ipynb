```python
# import libraries
import pandas as pd
import numpy as np
import missingno as msno
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
```

```python
# reading dataset
df = pd.read_csv("Social_Network_Ads.csv")
df
```

```python
# visualising null values
msno.matrix(df)
```

```python
# droping user id
df = df.drop(["User ID"], axis=1)
```

```python
df.head() # verifying dataset
```

```python
# Label Encoding Gender Values
label_encoder = LabelEncoder()
df["Gender"] = label_encoder.fit_transform(df["Gender"])
df
```

```python
# Dividing independent and dependent values of dataset
scaler = StandardScaler()

X = df.drop(["Purchased"], axis=1)
Y = df["Purchased"]

X = scaler.fit_transform(X)

```

```python
# splitting the training and testing data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state = 42)

linear_svc = LinearSVC(max_iter = 2000, random_state = 42)
svm_model = linear_svc.fit(x_train, y_train)

train_score = svm_model.score(x_test, y_test)
test_score = svm_model.score(x_train, y_train)

print("Train score: ", train_score)
print("Test score: ", test_score)
```
