```python
# importing libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("AirQualityUCI.csv")
df.head()
```

```python
# check NaN values for each feature
df.isnull().sum()
```

```python
# last 114 rows of the dataset are NaN values
df = df.drop(labels=range(9357, len(df)), axis=0)
```

```python
# last 2 columns are also filled with NaN values
df = df.drop(columns=['Unnamed: 15', 'Unnamed: 16'], axis=1)
```

```python
# Total Null values
df.isnull().sum()
```

```python
df.describe()
```

```python
df.head(5)
```

```python
# strip the time to hour from HH:MM:SS string format
df['Time'] = df['Time'].apply(lambda x: int(x.split(':')[0]))
```

```python
df
```

```python
# NHMC(GT) column lot of redudant false values
false_values = df['NMHC(GT)'][df['NMHC(GT)'] == -200].count()
print("False values with -200 :", false_values)

```

```python
# dropping the NMHC(GT) column
df = df.drop(columns=['NMHC(GT)'], axis=1)
df
```

```python
# Formating the date string to integer and labelling them by their day, month and year
df['Day']= df['Date'].apply(lambda x: int(x.split('/')[0]))
df['Month'] = df['Date'].apply(lambda x: int(x.split('/')[1]))
df['Year'] = df['Date'].apply(lambda x: int(x.split('/')[2]))
```

```python
# poping the columns and shifting their indices to front the dataset
day = df.pop('Day')
month = df.pop('Month')
year = df.pop('Year')
```

```python
df.insert(0, 'Day', day)
df.insert(1, 'Month', month)
df.insert(2, 'Year', year)
```

```python
# Now, drop the Date feature
df = df.drop(columns='Date', axis=1)
```

```python
df
```

```python
# checking pearson correlation with respect to RH (target) variable
df.corrwith(df['RH'], method='pearson')
```

```python
# seperating Independent variable and Dependent variable
X = df.drop(columns=['RH'], axis=1)
Y = df['RH']
```

```python
# spliting training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# training and testing linear regression model accuracy
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)

y_train_predict = linear_model.predict(x_train)
y_test_predict = linear_model.predict(x_test)

print("Linear Regression model")
print("Testing accuracy: ", r2_score(y_train_predict, y_train))
print("Training accuracy: ", r2_score(y_test_predict, y_test))
```

```python
# training and testing ridge regression model
ridge_model = Ridge()
ridge_model.fit(x_train, y_train)

y_train_predict = ridge_model.predict(x_train)
y_test_predict = ridge_model.predict(x_test)

print("Ridge Regression model")
print("Testing accuracy: ", r2_score(y_train_predict, y_train))
print("Training accuracy: ", r2_score(y_test_predict, y_test))
```

```python
# training and testing lasso regression model
lasso_model = Lasso(max_iter=2000)
lasso_model.fit(x_train, y_train)

y_train_predict = lasso_model.predict(x_train)
y_test_predict = lasso_model.predict(x_test)

print("Lasso model")
print("Testing accuracy: ", r2_score(y_train_predict, y_train))
print("Training accuracy: ", r2_score(y_test_predict, y_test))
```
